```python
echo "hello world" #returns the inputted text 
touch hello.txt
cat hello.txt 
echo "hello world" > hello.txt #overwrites text file with inputted text 
cat hello.txt 
echo "bye world" >> hello.txt #appends inputted text to text file
cat hello.txt 
```

```python
#in-class assignment: 
#1. get the current time using "date" function
#2. put the time (hh:mm:ss) inside of a "clock.txt" file in 20260223_shell
#3. make a folder "time", and move clock.txt into it

cd '/Users/cylam/Desktop/U of T PhD (2020-)/DSI Certificate/20260223_shell'
touch clock.txt
date "+%H:%M:%S" >> clock.txt #note how this works, but not echo date...
#that's because passing inputted text into target file is not a function of echo,
#it's a function of the > and >> 
#> and >> takes anything with an output (return) before > and passes it to 
#whatever is after >
cat clock.txt
mkdir time
mv clock.txt time
ls time
```
